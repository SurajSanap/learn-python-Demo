## Collecting all text files and create a csv
We have collected the review in the CSV files and need to convert into indivitual text files for further processing.

```python
import os
import glob
import pandas as pd
import fnmatch
```

```python
#List all text files
#'C:/Users/piyak/Github/my-project/Semi-supervised-sequence-learning-Project/imdb_review_scrapping/unsup'
directory = ('E:/00.Miscellaneous/~Macquarie University/Session 2/COMP8240 â€“ Applications of Data Science/Research papers/New dataset/IMDB/')
files_path = (directory + 'pos')
os.chdir(files_path)
all_filenames = fnmatch.filter(os.listdir(files_path), '*.txt')
#print(all_filenames)
```

```python
texts=[]
for file in all_filenames:
    #f = open(file,"r", encoding="utf-8")
    f = open(file,"r", encoding="ISO-8859-1")
    print(file)
    for text in file:
        lines=f.read()
        lines=lines.replace('[', "")
        lines=lines.replace(']', "")
        lines=lines.replace('"', "")
        lines=lines.split('\n')
        #print(lines)
        for line in lines:
            if line != '':
                #print(line)
                texts.append(lines) 
    f.close()
#Create df to sotre all texts
df = pd.DataFrame(texts) 
print(df.tail(5))
```

```python
#Converting to csv file
df.to_csv('pos.csv', index=False) 
```
